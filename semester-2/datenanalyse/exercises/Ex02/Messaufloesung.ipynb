Erstellen Sie ein Histogramm für jede Messung und plotten Sie es. Um das Histogramm zu berechnen, können Sie `np.histogram()`, und um es in der gewohnten Balken-Form zu plotten `ax.bar()` verwenden. Die Grösse der Bins hat einen grossen Einfluss auf das Histogramm. Wählen Sie verschiedene Bingrössen aus, die für kleine oder grosse Signalauflösungen optimiert sind. Was passiert, wenn Sie die Bingrösse kleiner als die Auflösung wählen?

Die Syntax der Funktionen finden Sie wieder mit der Hilfe Funktion: `np.histogram?` und `ax.bar?`. Alternativ können Sie auf die Hyperlinks am Anfang des Übungsblattes klicken, um zur Online-Dokumentation zu gelangen.

Um den Zusammenhang zwischen Signalauflösung und Bingrösse zu veranschaulichen, können Sie immer alle vier Histogramme mit derselben Bingrösse plotten, und dann verschiedene Werte für die Bingrösse ausprobieren.

```python
# A histogram reveals the distribution of measurements, where the voltages tend to be
# The bin size must be significantly larger than the resolution, otherwise there may only be one value in each bin
# revealing very little about the distribution
def calc_res(data):
    # Resolution = smallest difference between measurements
    deltas = np.diff(np.sort(data))
    return round(np.min(deltas[np.nonzero(deltas)]), 4)

def calc_bin_width(data, bins):
    d_range = np.max(data) - np.min(data)
    return round(d_range / bins, 4)

fig, ax = plt.subplots()
for i in [0]:
    # comments='#' by default
    data = np.loadtxt(f"Aufgabe1/measurement_{i +1}.txt", delimiter=',')
    res = calc_res(data[:,1])
    # This is a good balance
    bin_w = round(res*1.1, 4)
    bins = round((np.max(data[:, 1]) - np.min(data[:, 1])) / bin_w)
    ax.hist(data[:, 1], bins, label=f"meas={i+1}, res={res}, bin_w={bin_w}")
    
ax.set_xlabel('Voltage (mV)')
ax.set_ylabel(r'Frequency Density ($V^{-1}$)')
ax.set_title("Histogram") 
ax.legend()

plt.show()
```

```python
# Ihr Code
def calc_res(data):
    # Resolution = smallest difference between measurements
    deltas = np.diff(np.sort(data))
    return round(np.min(deltas[np.nonzero(deltas)]), 4)

data = np.loadtxt("Aufgabe2/osc_data_1.txt", delimiter=',')
samp_rate = 1/ calc_res(data[:, 0])
print(f"Sampling rate: {samp_rate} Hz")

fig, ax = plt.subplots()
ax.plot(data[:, 0], data[:, 1], '.-', label=f"samp_rate={samp_rate}")
    
ax.set_xlabel('time (s)')
ax.set_ylabel('voltage (V)')
ax.set_title("Oscilloscope") 
ax.legend()

# Zoom to smaller time interval
ax.set(xlim=(0, 0.1))
```
